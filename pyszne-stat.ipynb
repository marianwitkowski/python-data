```python
df = pd.read_csv("data/pyszne-dt.csv", usecols=[1,2,3])
df
```

```python
df = df.sort_values('ts', ascending=True).reset_index(drop=True)
```

```python
df.deliv_time = df.deliv_time.str.strip('min')
```

```python
df.deliv_time = df.deliv_time.astype('int')
```

```python
df['ts1'] = df.ts.str[:10]
```

```python
fig = px.box(df, x="ts", y="deliv_time")
fig.show()
```
