```
import pandas as pd
from pandas_profiling import ProfileReport
```

```
df = pd.read_csv("http://hackingpython.pl/datasets/otomoto.csv")
```

```
df.sample(n=120)
```

```
profile = ProfileReport(df, title="Otomoto.pl Profiling Report")
```

```
profile.to_notebook_iframe()
```

```
profile.to_file("otomoto_report.html")
```
