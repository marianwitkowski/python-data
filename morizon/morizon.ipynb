# Analiza danych ogłoszeń morizon.pl

Warszawa, mieszkania, cena: do 1mln zł

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_1mln_wwa = pd.read_csv("morizon-wwa-1mln.csv")
```

## Histogram dla ceny mieszkania

```python
df_1mln_wwa.Cena.plot(kind='hist', figsize=(12,7), bins=50, color='red', grid=True, title='Ceny mieszkań')
```

## Histogram dla ceny m2

```python
df_1mln_wwa.Cena_m2.plot(kind='hist', figsize=(12,7), bins=50, color='red', grid=True, title='Ceny m2')
```

## Histogram dla powierzchni mieszkania

```python
df_1mln_wwa.Powierzchnia.plot(figsize=(12,7), bins=50, color='red', kind='hist', grid=True, title='Powierzchnia')
```

## Liczba ogłoszeń wg dzielnic

```python
dfg = df_1mln_wwa.groupby('Dzielnica')
dfg.size().nlargest(15).sort_values(ascending=True).plot(kind='barh', figsize=(10,10), grid=True, color='red', title='Dzielnice')
```

## Rozkład liczby pokojów

```python
dfg = df_1mln_wwa.groupby('Pokoje')
dfg.size().nlargest(5).sort_values(ascending=True).plot(kind='barh', figsize=(10,10), grid=True, color='red', title='Liczba pokojów')
```

## Deweloperzy zgłaszający inwestycje

```python
dfg = df_1mln_wwa.groupby('Deweloper')
dfg.size().nlargest(15).sort_values(ascending=True).\
    plot(kind='barh', figsize=(10,10), grid=True, color='red', title='Inwestorzy')
```

## Rok budowy

```python
df_1mln_wwa[df_1mln_wwa["Rok budowy"]>=1900]["Rok budowy"].plot(kind='hist', 
                                                                title='Rok budowy', grid=True, 
                                                                figsize=(12,7), bins=40, color='red')
```

## Rozkład wartości wysokości pomieszczeń

```python
df_1mln_wwa['Wysokość wnętrza'].plot(kind='hist', title='Wysokość pomieszczeń', grid=True, figsize=(12,7), bins=30, color='red')
```

## Typ budynku

```python
dfg = df_1mln_wwa.groupby('Typ budynku')
dfg.size().nlargest(10).sort_values(ascending=True).plot(kind='barh', figsize=(10,10), grid=True, color='red', title='Typ budynku')
```

## Liczba pięter w budynku

```python
dfg = df_1mln_wwa.groupby('Liczba pięter')
dfg.size().nlargest(15).sort_values(ascending=True).plot(kind='barh', title='Liczba pięter budynku', 
                                                         figsize=(10,10), grid=True, color='red')
```

# Powierzchnia łazienek

```python
dfg = df_1mln_wwa[df_1mln_wwa["Powierzchnia łazienki"]>0].groupby('Powierzchnia łazienki')
dfg.size().nlargest(15).sort_values(ascending=True).plot(title='Pow. łazienki', kind='barh', 
                                                         figsize=(10,10), grid=True, color='red')
```
